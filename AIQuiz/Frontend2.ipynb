```python
from flask import Flask, render_template_string
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

app = Flask(__name__)

df = None  # Define a global variable to store the DataFrame

def read_data():
    global df
    df = pd.read_csv('frauddataset.csv')

@app.route('/')
def home():
    return '''
    <html>
    <head>
    <title>Bank Fraud Detection</title>
    <style>
    body {
    display: flex;
    align-items: center;
    justify-content: center;
    flex-direction: column;
    font-family: Arial, sans-serif;
    background-color: #f7f7f7;
    }


    .container {
        max-width: 960px;
        margin: 0 auto;
        padding: 20px;
    }

    /* Title styles */
    .title {
        color: #333;
        font-size: 32px;
        margin-bottom: 20px;
        text-align: center;
    }

    /* Menu styles */
    .menu {
        list-style-type: none;
        margin: 0;
        padding: 0;
    }

    .menu li {
        margin-bottom: 10px;
    }

    .menu li a {
        display: inline-block;
        padding: 10px 20px;
        background-color: #4c75af;
        border-radius: 5px;
        color: #fff;
        text-decoration: none;
        transition: background-color 0.3s ease;
    }

    .menu li a:hover {
        background-color: #2c4e8a;
    }

    /* Button styles */
    .btn {
        display: inline-block;
        padding: 10px 20px;
        background-color: #4c75af;
        border-radius: 5px;
        color: #fff;
        text-decoration: none;
        transition: background-color 0.3s ease;
    }

    .btn:hover {
        background-color: #2c4e8a;
    }

    /* Table styles */
    .table {
        width: 100%;
        border-collapse: collapse;
        border: 1px solid #ddd;
    }

    .table th, .table td {
        padding: 8px;
        border-bottom: 1px solid #ddd;
    }

    .table th {
        background-color: #4c75af;
        color: #fff;
    }

    /* Graph container styles */
    .graph-container {
        margin-top: 20px;
    }

    /* Graph styles */
    .graph {
        width: 100%;
        height: 400px;
    }
   </style>
   </head>
    
    <h1 class="title">Bank Fraud Detection</h1>
    <body>
    <ul class="menu">
        <li><a class="link" href="/read">Read Data</a></li>
        <li><a class="link" href="/start">Show Start</a></li>
        <li><a class="link" href="/end">Show End</a></li>
        <li><a class="link" href="/info">Show Info</a></li>
        <li><a class="link" href="/null_vals">Show Null Values</a></li>
        <li><a class="link" href="/graph">Show Graph</a></li>
        <li><a class="link" href="/plot">Plot Scatter</a></li>
        <li><a class="link" href="/plot_face">Plot Facet</a></li>
        <li><a class="link" href="/plot_line">Plot Line</a></li>
        <li><a class="link" href="/plot_box">Plot Box</a></li>
    </ul>
    </body>
    </html>
    '''

@app.route('/read')
def read():
    read_data()
    return "Data loaded successfully!"

@app.route('/start')
def show_start():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        return render_template_string(df.head().to_html())

@app.route('/end')
def show_end():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        return render_template_string(df.tail().to_html())

@app.route('/info')
def show_info():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        return df.info()

@app.route('/null_vals')
def show_null_vals():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        return render_template_string(df.isnull().sum().to_html())

@app.route('/graph')
def show_graph():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        df_fraud = df.loc[df.fraud == 1]
        df_non_fraud = df.loc[df.fraud == 0]
        sns.countplot(x='fraud', data=df)
        plt.title("Count of Fraudulent Payments")
        plt.show()
        return f"Number of normal examples: {df_non_fraud.fraud.count()}<br>Number of fraudulent examples: {df_fraud.fraud.count()}"

@app.route('/plot')
def plot_scatter():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        fig = px.scatter(df, x="category", y="amount")
        return fig.to_html()

@app.route('/plot_face')
def plot_facet():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        fig = px.scatter(df, x="fraud", y="amount", facet_row='gender')
        return fig.to_html()

@app.route('/plot_line')
def plot_line():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        fig = px.line(df, x="step", y="amount", title="Transaction Amount By numerous steps")
        return fig.to_html()

@app.route('/plot_box')
def plot_box():
    if df is None:
        return "Data not loaded. Please click 'Read Data' first."
    else:
        fig = px.box(df, y='age')
        return fig.to_html()
    


if __name__ == '__main__':
    app.run()
```
