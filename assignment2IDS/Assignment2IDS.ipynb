```python
import numpy as np
array = np.zeros((5, 5), dtype=int)
num = 1
top, bottom, left, right = 0, 5, 0, 5

while num <= 25:
    for i in range(left, right):
        array[top, i] = num
        num += 1
    top += 1

    for i in range(top, bottom):
        array[i, right - 1] = num
        num += 1
    right -= 1

    for i in range(right - 1, left - 1, -1):
        array[bottom - 1, i] = num
        num += 1
    bottom -= 1

    for i in range(bottom - 1, top - 1, -1):
        array[i, left] = num
        num += 1
    left += 1


print(array)

subarray = np.zeros((3, 3), dtype=int)
for i in range(1, 4):
    for j in range(1, 4):
        subarray[i - 1, j - 1] = array[i, j]
print("3x3 subarray from the center of the original array")
print(subarray)
new_array=array[::-1]
print("Reversed Array")
print(new_array)
indices=[]

for i,row in enumerate(new_array):
    for y,val in enumerate(row):
         if val > 15:
            indices.append((i,y,val))
print("Indices of all elements greater than 15")
print(indices)
reshaped_array=[]
for row in array:
    for val in row:
        reshaped_array.append(val)
print(reshaped_array)
```

```python
def check_winner(board):
    for i in range(10):
        if np.all(board[i, :] == 1) or np.all(board[:, i] == 1):
            return 1  # Player 1 wins

        if np.all(board[i, :] == 2) or np.all(board[:, i] == 2):
            return 2  # Player 2 wins

    if np.all(np.diag(board) == 1) or np.all(np.diag(np.fliplr(board)) == 1):
        return 1  # Player 1 wins

    if np.all(np.diag(board) == 2) or np.all(np.diag(np.fliplr(board)) == 2):
        return 2  # Player 2 wins

    if np.all(board != 0):
        return 0  # Tie

    # Game still in progress
    return -1

winner = check_winner(game_board_scenario_1)
print("Winner:", winner)
winner = check_winner(game_board_scenario_2)
print("Winner:", winner)
winner = check_winner(game_board_scenario_3)
print("Winner:", winner)
```
